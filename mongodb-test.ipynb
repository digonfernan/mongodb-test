# Biblioteca

```python
import pymongo
import os
import pandas as pd
import json
```

# Conexão com o MongoDB

```python
# Cria conexão local
client = pymongo.MongoClient('mongodb://localhost:27017')
print(client)
```

# Carregando o dataset

```python
# Import dataframe
whrdataset = pd.read_csv('https://raw.githubusercontent.com/digonfernan/mongodb-test/main/world-happiness-report-2021.csv')
```

```python
# Visualiza os dados
display(whrdataset.shape)
whrdataset.head()
```

# Preparando para o export

```python
# Cria um dicionário
whrdic = whrdataset.to_dict(orient = 'records')
```

```python
# Visualiza o dicionário
whrdic
```

# Exportando para o MongoDB

```python
# Cria o database
whrdb = client['whrtotal']
```

```python
# Visualiza o database
print(whrdb)
```

```python
# Export o database
# db.Iris.insert_many(data)
whrdb.whrdb21.insert_many(whrdic)
```
